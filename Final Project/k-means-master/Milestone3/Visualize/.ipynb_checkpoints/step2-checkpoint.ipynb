```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('http://statistical-research.com/wp-content/uploads/2013/11/sample_geo.txt', sep="\t")
```

```python
print(data[:10])
```

```python
fig, ax = plt.subplots()
ax.scatter(data['Longitude'], data['Latitude'], alpha=0.3)
ax.set_ylim((20,55))
ax.set_xlim((-130,-60))
ax.set_xlabel('Long', fontsize=10)
ax.set_ylabel('Lat', fontsize=10)
ax.set_title('simple geo')
ax.grid(True)
fig.tight_layout()

plt.show()
```
