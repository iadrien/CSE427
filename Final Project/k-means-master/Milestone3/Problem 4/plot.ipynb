```python
import ast
from numpy import array
from os.path import join
from matplotlib import pyplot as plt
```

```python
def read_result(p):
    result = []
    with open(join(p, 'part-00000'), 'r') as f:
        result += [ast.literal_eval(line) for line in f]
    with open(join(p, 'part-00001'), 'r') as f:
        result += [ast.literal_eval(line) for line in f]
    return result
```

```python
clusters_euclidean_3 = read_result('./dft-accident-data/clusters_euclidean_3')
centers_euclidean = [array([-0.13638543, 51.697445  ]), array([-2.48161016, 51.78915769]), array([-2.32147019, 54.23874128])]
```

```python
%%time
fig, ax = plt.subplots(figsize=(15, 10))
for c in clusters_euclidean_3:
    ax.scatter(*zip(*c[1]) ,alpha=0.3)
    ax.scatter(centers_euclidean[c[0]][0],centers_euclidean[c[0]][1], marker="X",alpha=1)
ax.set_xlabel('Lat', fontsize=10)
ax.set_ylabel('Long', fontsize=10)
ax.set_title('accident data with euclidean distance k=3')
ax.grid(True)
fig.tight_layout()

plt.show()
```
